```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

df = pd.read_csv('canada_per_capita_income.csv')
df

# Commented out IPython magic to ensure Python compatibility.
# %matplotlib inline
plt.xlabel('year')
plt.ylabel('Income')
plt.scatter(df.year, df.income,color='red',marker='+')

new_df = df.drop('income',axis='columns')
new_df

income = df.income
income

# # Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,income)

reg.predict([[2020]])
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
df3=pd.read_csv('hiring.csv')
df3.isnull().sum()
df3['experience'] = df3["experience"].fillna(df3['experience'].mode())
df3
df3.experience = df3.experience.astype(str)
def convert_to_int(word):
    word_dict = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5, 'six':6, 'seven':7, 'eight':8,
                'nine':9, 'ten':10, 'eleven':11, 'twelve':12, 'zero':0, 0:0}
    return word_dict[word]
df3.experience=df3.experience.apply(lambda x : convert_to_int(x))
print(df3.head(10))
df3.fillna(df3.mean(),inplace=True)
print(df3.isnull().sum())
print(df3.head(10))
reg=linear_model.LinearRegression()
reg.fit(df3.drop('salary($)',axis='columns'),df3['salary($)'])
print(reg.predict([[2,9,6]]))
print(reg.predict([[12,10,10]]))
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.preprocessing import OrdinalEncoder
df3=pd.read_csv('1000_Companies.csv')
df3.isnull().sum()

ordinal_encoder = OrdinalEncoder(categories=[["California", "Florida", "New York"]])
df3["State"] = ordinal_encoder.fit_transform(df3[["State"]])
df3

reg=linear_model.LinearRegression()
reg.fit(df3.drop('Profit',axis='columns'),df3['Profit'])

print(reg.predict([[91694.48,515841.3,11931.24, 1]]))
```

```python
# -*- coding: utf-8 -*-
"""Linear-Regression-Housing_Area_Price.ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1CAlZml-P6V2V1RIrodgMfF8L3Ux4V9FT
"""

import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

df = pd.read_csv('housing_area_price.csv')
df

# Commented out IPython magic to ensure Python compatibility.
# %matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')

new_df = df.drop('price',axis='columns')
new_df

price = df.price
price

# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,price)

"""(1) Predict price of a home with area = 3300 sqr ft"""

reg.predict([[3300]])

reg.coef_

reg.intercept_

"""Y = m * X + b (m is coefficient and b is intercept)"""

3300*135.78767123 + 180616.43835616432

"""(1) Predict price of a home with area = 5000 sqr ft"""

reg.predict([[5000]])
```

```python
# -*- coding: utf-8 -*-
"""Multiple_LR_HomePrice.ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1fK78C8TPV44HdvT6lsMhaau2wMtKXquQ
"""

import pandas as pd
import numpy as np
from sklearn import linear_model

df = pd.read_csv('homeprices_Multiple_LR.csv')
df

"""Data Preprocessing: Fill NA values with median value of a column"""

df.bedrooms.median()

df.bedrooms = df.bedrooms.fillna(df.bedrooms.median())
df

reg = linear_model.LinearRegression()
reg.fit(df.drop('price',axis='columns'),df.price)

reg.coef_

reg.intercept_

"""Find price of home with 3000 sqr ft area, 3 bedrooms, 40 year old"""

reg.predict([[3000, 3, 40]])

112.06244194*3000 + 23388.88007794*3 + -3231.71790863*40 + 221323.00186540384
```
